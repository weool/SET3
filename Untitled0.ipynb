```python
import matplotlib.pyplot as plt
import numpy as np
import math
def read_data(filename):
    points = []
    areas = []
    with open(filename, 'r') as f:
        for line in f:
            n, area = map(float, line.split())
            points.append(n)
            areas.append(area)
    return points, areas
points1, areas1 = read_data('data_1.txt')
points2, areas2 = read_data('data_2.txt')
def calculate_exact_area():
    S_triangle = 0.5
    S_segment1 = (math.pi/2 - math.sin(math.pi/2)) / 2 * 1**2
    angle = math.asin(0.8)
    r = math.sqrt(5)/2
    S_segment2 = (angle - math.sin(angle)) / 2 * r**2

    total_area = S_triangle + S_segment1 + 2 * S_segment2
    return total_area

exact_area = calculate_exact_area()
plt.subplot(2, 2, 1)
plt.plot(points1, areas1, 'b-', label=' 1 ( )', linewidth=1.5, alpha=0.8)
plt.plot(points2, areas2, 'r-', label=' 2 ( )', linewidth=1.5, alpha=0.8)
plt.axhline(y=exact_area, color='g', linestyle='--', linewidth=2, label=f'  ({exact_area:.4f})')
plt.xlabel(' , N')
plt.ylabel(' ')
plt.title('     ')
plt.legend()
plt.grid(True, alpha=0.3)

plt.figure(figsize=(12, 8))

#     
relative_error1 = [abs(area - exact_area) / exact_area * 100 for area in areas1]
relative_error2 = [abs(area - exact_area) / exact_area * 100 for area in areas2]

plt.subplot(2, 2, 1)
plt.plot(points1, relative_error1, 'b-', label=' 1 ( )', linewidth=1.5, alpha=0.8)
plt.plot(points2, relative_error2, 'r-', label=' 2 ( )', linewidth=1.5, alpha=0.8)
plt.xlabel(' , N')
plt.ylabel(' , %')
plt.title('     ')
plt.legend()
plt.grid(True, alpha=0.3)
```
